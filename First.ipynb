```python
#image Embossing
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread('1.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

k1=np.array([[0,-1,-1],[1,0,-1],[1,1,0]])
k2=np.array([[-1,-1,0],[-1,0,1],[0,1,1]])
k3=np.array([[1,0,0],[0,0,0],[0,0,-1]])
k4=np.array([[0,0,1],[0,0,0],[-1,0,0]])

img1=cv2.filter2D(img,-1,k1)
img2=cv2.filter2D(img,-1,k2)
img3=cv2.filter2D(img,-1,k3)
img4=cv2.filter2D(img,-1,k4)

plt.subplot(2,3,1),plt.imshow(img),plt.title('source img'),plt.axis('off')
plt.subplot(2,3,2),plt.imshow(img1,cmap='gray'),plt.title('k1'),plt.axis('off')
plt.subplot(2,3,3),plt.imshow(img2,cmap='gray'),plt.title('k2'),plt.axis('off')
plt.subplot(2,3,4),plt.imshow(img3,cmap='gray'),plt.title('k3'),plt.axis('off')
plt.subplot(2,3,5),plt.imshow(img4,cmap='gray'),plt.title('k4'),plt.axis('off')
```

```python
#Erosion and Dilation
import cv2
import numpy as np
import matplotlib.pyplot as plt

#Load the image
img=cv2.imread('4.jpg')
img=cv2.bitwise_not(img)

#Define a kernal
k1=np.ones((3,3),np.uint8)

#Erosion
erosion=cv2.erode(img,k1,iterations=1)

#Dilation
dilation=cv2.dilate(img,k1,iterations=1)

#Display the image
plt.subplot(1,3,1),plt.imshow(img,cmap='gray'),plt.title('source'),plt.axis('off')
plt.subplot(1,3,2),plt.imshow(erosion,cmap='gray'),plt.title('erosion'),plt.axis('off')
plt.subplot(1,3,3),plt.imshow(dilation,cmap='gray'),plt.title('dilation'),plt.axis('off')

# cv2.imshow('source',img)
# cv2.imshow('Erosion',erosion)
# cv2.imshow('Dilation',dilation)
# cv2.waitkey(0)
# cv2.destroyAllwindows()



```

```python
#contrasting the color image
#using the global Histogram Equivalence(GHE)
import cv2
import matplotlib.pyplot as plt
import numpy as np

#Loading the source image
img=cv2.imread('10.jpg')

#converting the image from BGR to Ycrcb
ycrcb=cv2.cvtColor(img,cv2.COLOR_BGR2YCrCb)
#splite the elements of the ycrcb
y,cr,cb=cv2.split(ycrcb)

#Applying the contrast using Global Histogram Equalization
y_eq=cv2.equalizeHist(y)
ycrcb_eq=cv2.merge([y_eq,cr,cb])

#converting the color from YCrCb to BGR
global_eq=cv2.cvtColor(ycrcb_eq,cv2.COLOR_YCrCb2BGR)
#converting the color from BGR to RGB
global_eq1=cv2.cvtColor(global_eq,cv2.COLOR_BGR2RGB)
source=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#ploting the image
plt.subplot(1,2,1),plt.imshow(source,cmap='gray'),plt.title('source'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(global_eq1,cmap='gray'),plt.title('Contrast img using GHE'),plt.axis('off')
```

```python
#contrasting the color image
#using the global Histogram Equivalence(GHE)
import cv2
import matplotlib.pyplot as plt
import numpy as np

#Loading the source image
img=cv2.imread('10.jpg')

#converting the image from BGR to Ycrcb
ycrcb=cv2.cvtColor(img,cv2.COLOR_BGR2YCrCb)
#splite the elements of the ycrcb
y,cr,cb=cv2.split(ycrcb)

#Applying the contrast using CLAHE
clahe=cv2.createCLAHE(clipLimit=2.0,tileGridSize=(8,8))
y_clahe=clahe.apply(y)
ycrcb_clahe=cv2.merge([y_clahe,cr,cb])

#converting the color from ycrcb to bgr
global_clahe=cv2.cvtColor(ycrcb_clahe,cv2.COLOR_YCrCb2BGR)
#converting the color from BGR to RGB
global_clahe=cv2.cvtColor(global_clahe,cv2.COLOR_BGR2RGB)
source=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)


#ploting the image
plt.subplot(1,2,1),plt.imshow(source,cmap='gray'),plt.title('source'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(global_clahe,cmap='gray'),plt.title('Contrast img using CLAHE'),plt.axis('off')
```

```python
#Applying advanced image filters
#1).Bilateral(for advanced smoothing affect)
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('10.jpg')
bilateral=cv2.bilateralFilter(img,d=200,sigmaColor=100,sigmaSpace=100)

plt.subplot(1,2,1),plt.imshow(img),plt.title('source'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(bilateral),plt.title('smooth image using Bilateral'),plt.axis('off')
```

```python
#Applying advanced image filters
#2).unsharp(for sharpening the image)
import cv2
import matplotlib.pyplot as plt

img=cv2.imread('mountain.jpg')
#blur
gauss=cv2.GaussianBlur(img,(9,9),10)

#unsharp masking
unsharp=cv2.addWeighted(img,1.5,gauss,-0.5,0)

plt.subplot(1,2,1),plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB)),plt.title('source'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(cv2.cvtColor(unsharp,cv2.COLOR_BGR2RGB)),plt.title('sharping using unsharp'),plt.axis('off')
```

```python
#3).scharr(for edge detection)
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('tiger.jpg',0)

#Apply scharr filter
scharrx=cv2.Scharr(img,cv2.CV_64F,1,0)#vertical edges
scharry=cv2.Scharr(img,cv2.CV_64F,0,1)#horizontal edges

#convert to uint8 for display
scharrx=cv2.convertScaleAbs(scharrx)
scharry=cv2.convertScaleAbs(scharry)

#convert to uint8 for display
scharrx=cv2.convertScaleAbs(scharrx)
scharry=cv2.convertScaleAbs(scharry)

#combine both direction
scharr_combined=cv2.addWeighted(scharrx,0.5,scharry,0.2,0)
#scharr_combined=cv2.magnitude(scharrx,scharry)
#plot the image
plt.figure(figsize=(5,5))
plt.subplot(2,2,1),plt.imshow(img,cmap='gray'),plt.title('source'),plt.axis('off')
plt.subplot(2,2,2),plt.imshow(scharrx,cmap='gray'),plt.title('scharrx'),plt.axis('off')
plt.subplot(2,2,3),plt.imshow(scharry,cmap='gray'),plt.title('scharry'),plt.axis('off')
plt.subplot(2,2,4),plt.imshow(scharr_combined,cmap='gray'),plt.title('scharr'),plt.axis('off')


```

```python
#Object detection from the image
#Face
import cv2
#Load pre=trained face detector (Haar Cascade)
face_cascade=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
#Load image
img=cv2.imread('face1.jpg')
gray_scale=(cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)) 
#Detect faces
faces=face_cascade.detectMultiScale(gray_scale,scaleFactor=1.1,minNeighbors=4)
print(faces)
#Draw rectangles around faces
for(x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),3)
cv2.imshow('Face Detector',img)
cv2.waitKey()
cv2.destroyAllWindows()


```

```python
#object detection using camera
#face
import cv2
#Load pre-trained face detector (Haar Cascade)
face_cascade=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
#open the camera(0 for inbuilt camera,1 for external camera)
cap=cv2.VideoCapture(0)
#capture from the camera
while True:
    #capture frame by frame
    ret,frame=cap.read()
    if not ret:
        break #if camera not working
gray=(cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)) #Haar cascade works on single channel color.
#Detect faces
faces=face_cascade.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=4)
print(faces)
#Draw rectangle around faces
for(x,y,w,h) in faces:
    cv2.rectangle(frame,(x,y),(x+w,y+h),(0,0,255),1)
cv2.imshow('Face Detection from camera',frame)
#close the camera when Esc key is pressed
if cv2.waitKey(1)==27:
    break
cv2.destroyAllWindows()
```

```python
#Applying the sentence takenization on the text
from nltk.tokenize import sent_tokenize
#input text
text="I love NLP. It's fun, isn't it?"
text1="I love NLP.It's fun, isn't it?"
#sentence tokenisation
tokenised_sentence=sent_tokenize(text)
print('sentence tokeniser:',tokenised_sentence)
tokenised_sentence1=sent_tokenize(text1)
print('sentence tokeniser:',tokenised_sentence1)
```

```python
#Applying the word tokenization on the text
from nltk.tokenize import word_tokenize
#input text
text="I love NLP. It's fun, isn't it?"
#word tokenisation
tokenised_word=word_tokenize(text)
print('word tokeniser:',tokenised_word)
```

```python
#Applying the word Punctuation tokenization on the text
from nltk.tokenize import wordpunct_tokenize
#input text
text="I love NLP. It's fun, isn't it?"
#word tokenisation
tokenised_wordpunct=wordpunct_tokenize(text)
print('word tokeniser:',tokenised_wordpunct)
```

```python
#Stemming all 3 in one
from nltk.stem import PorterStemmer, SnowballStemmer, LancasterStemmer
#creating the object of stemmer
porter=PorterStemmer()
snowball=SnowballStemmer('english')
lancaster=LancasterStemmer()
#input words
words=["student","water","studies","studying","happiness","playing","played"]
print('Porter Stemmer')
for w in words:
    print(w,'-->',porter.stem(w))
print('\nSnowball Stemmer\n')
for w in words:
    print(w,'-->',snowball.stem(w))
print('\nLancaster Stemmer\n')
for w in words:
    print(w,'-->',lancaster.stem(w))
```
